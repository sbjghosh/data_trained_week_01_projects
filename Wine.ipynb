```python
import pandas as pd
import numpy as np
import dtale
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```python
red=pd.read_csv('winequality-red.csv',sep=';')
white=pd.read_csv('winequality-white.csv',sep=';')
```

```python
red.head()
```

```python
white.head()
```

```python
red.columns = [x.strip().replace(' ','_') for x in red.columns]
red.head()
```

```python
white.columns = [x.strip().replace(' ','_') for x in white.columns]
white.head()
```

```python
red['is_red']=1
```

```python
red.head()
```

```python
white['is_red']=0
white.head()
```

```python
df=pd.concat([red,white],ignore_index=True)
df.describe()
```

```python
dtale.show(red)
```

```python
X=red.drop('quality',axis=1)
```

```python
X.head()
```

```python
y=red['quality']
y.head()
```

```python
from sklearn import model_selection
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
X_train,X_test,y_train,y_test=model_selection.train_test_split(X,y,test_size=0.25)
```

```python
linear_reg_pipeline=Pipeline([('Scalar',StandardScaler()),('PCA',PCA(n_components=4)),('Linear_Regression',LogisticRegression())])
linear_reg_pipeline.fit(X_train,y_train)
linear_reg_pipeline.score(X_test,y_test)
```

```python
random_forest_pipeline=Pipeline([('Scalar',StandardScaler()),('PCA',PCA(n_components=5)),('Random_forest',RandomForestClassifier(n_estimators=25))])
random_forest_pipeline.fit(X_train,y_train)
random_forest_pipeline.score(X_test,y_test)
```

```python
poly = PolynomialFeatures(degree = 2, interaction_only=False, include_bias=False)

lr = LogisticRegression(max_iter=1000)

pipe = Pipeline([('polynomial_features',poly), ('Scalar',StandardScaler()),('PCA',PCA(n_components=32)),('logistic_regression',lr)])
pipe.fit(X_train, y_train)
pipe.score(X_test, y_test)
```

```python
poly = PolynomialFeatures(degree = 2, interaction_only=False, include_bias=False)

lr = LogisticRegression(max_iter=1000)

pipe = Pipeline([('polynomial_features',poly), ('Scalar',StandardScaler()),('PCA',PCA(n_components=32)),('Random_forest',RandomForestClassifier(n_estimators=25))])
pipe.fit(X_train, y_train)
pipe.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
rf=RandomForestClassifier(n_estimators=100,random_state=42)
rf.fit(X_train,y_train)
predrf=rf.predict(X_test)
print(metrics.accuracy_score(predrf,y_test))
print(metrics.confusion_matrix(predrf,y_test))
print(metrics.classification_report(predrf,y_test))
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn import metrics
ad=AdaBoostClassifier(n_estimators=100,random_state=42)
ad.fit(X_train,y_train)
predrf=ad.predict(X_test)
print(metrics.accuracy_score(predrf,y_test))
print(metrics.confusion_matrix(predrf,y_test))
print(metrics.classification_report(predrf,y_test))
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
sv=SVC()
from sklearn import metrics
ad=AdaBoostClassifier(base_estimator=sv,n_estimators=100,random_state=42,algorithm='SAMME')
ad.fit(X_train,y_train)
predrf=ad.predict(X_test)
print(metrics.accuracy_score(predrf,y_test))
print(metrics.confusion_matrix(predrf,y_test))
print(metrics.classification_report(predrf,y_test))
```
